```python
# Import biblioteki PyGame
import pygame
from pygame.locals import *

# Zainicjowanie gry w PyGame
pygame.init()

# Ustawienie screena
screen = pygame.display.set_mode((1920, 1080))

# Kontroler FPS-ów
clock = pygame.time.Clock()

# Pętla gry
run = True
while run:

    # 4 polecenia pozwalające na NORMALNE wyłączenie gry.
    for event in pygame.event.get():
        if event.type == QUIT:
            run = False
            break

    # Wypełnianie ekranu kolorem
    screen.fill((100, 100, 100))

    # Rysowanie obiektu
    pygame.draw.rect(screen, (255, 0, 0), Rect(300, 400, 350, 550))

    # Aktualizacja zawartości ekranu
    pygame.display.flip()

    # Czekanie 1/60 sekundy od poprzedniego wywołania clock.tick()
    clock.tick(60)

# Wyłączenie PyGame
pygame.quit()
```

## 2.1. Dodanie sterowania

```python
# Import biblioteki PyGame
import pygame
from pygame.locals import *

# Zainicjowanie gry w PyGame
pygame.init()

# Ustawienie screena
screen = pygame.display.set_mode((1920, 1080))

# Kontroler FPS-ów
clock = pygame.time.Clock()

# GRACZ
x = 200
y = 200

velocity = 5



# Pętla gry
run = True
while run:

    # 4 polecenia pozwalające na NORMALNE wyłączenie gry.
    for event in pygame.event.get():
        if event.type == QUIT:
            run = False
            break

    # ========= LOGIKA GRY ========= #
    
    # Przechwytywanie wciśniętych klawiszy
    keys = pygame.key.get_pressed()

    if keys[K_d]:
        x += velocity
    if keys[K_a]:
        x -= velocity
    if keys[K_w]:
        y -= velocity
    if keys[K_s]:
        y += velocity

    # ========= RENDEROWANIE ========= #
    # Wypełnianie ekranu kolorem
    screen.fill((0, 0, 0))

    # Rysowanie obiektu
    pygame.draw.rect(screen, (255, 0, 0), Rect(x, y, 150, 150))

    # Aktualizacja zawartości ekranu
    pygame.display.flip()

    # Czekanie 1/60 sekundy od poprzedniego wywołania clock.tick()
    clock.tick(60)

# Wyłączenie PyGame
pygame.quit()
```

# 2.2. Refactoring - dla lepszej czytelności projektu:

```python
# Potrzebne importy
import pygame
from pygame.locals import *

# Zmienne rozdzielczości ekranu
SCREEN_WIDTH = 1920
SCREEN_HEIGHT = 1080

# Kolory
COLOR_RED = (255, 0, 0)
COLOR_WHITE = (255, 255, 255)
COLOR_BLACK = (0, 0, 0)

# Inicjalizacja wszystkich mechanizmów pythona (WAŻNE);
pygame.init()

# Parametry Screena
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Zegar kontrolujący FPS-y
clock = pygame.time.Clock()

# GRACZ
x = 100
y = 100
velocity = 5
width = 150
height = 150

# Pętla gry
running = True
while running:

    # =============== ZAMYKANIE GRY ================= #
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            break


    # =============== STEROWANIE ================= #    
        
    keys = pygame.key.get_pressed()

    if keys[K_d]:
        x += velocity
    if keys[K_a]:
        x -= velocity
    if keys[K_w]:
        y -= velocity
    if keys[K_s]:
        y += velocity


    # ======= OGRANICZENIE RUCHU GRACZA ======= #

    if x < 0:
        x = 0

    if x > SCREEN_WIDTH - width:
        x = SCREEN_WIDTH - width

    if y < 0:
        y = 0

    if y > SCREEN_HEIGHT - height:
        y = SCREEN_HEIGHT - height


    # =============== RENDEROWANIE ================= #

    # Wypełnianie ekranu kolorem
    screen.fill(COLOR_BLACK)

    # Rysowanie obiektu
    pygame.draw.rect(screen, COLOR_RED, Rect(x, y, width, height))
    pygame.draw.circle(screen, COLOR_WHITE, (x, y), 3)
    # Aktualizacja zawartości ekranu
    pygame.display.flip()


    # Aktualizacja gry
    pygame.display.flip()

    # Czekanie na kolejną klatkę
    clock.tick(60)

pygame.quit()
```

## 2.3. Fizyka gry

```python
# Import biblioteki PyGame
import pygame
from pygame.locals import *

SCREEN_WIDTH = 1920
SCREEN_HEIGHT = 1080
FPS = 60

COLOR_RED = (255, 0, 0)
COLOR_WHITE = (255, 255, 255)
COLOR_BLACK = (0, 0, 0)

# Zainicjowanie gry w PyGame
pygame.init()



# Ustawienie screena
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Kontroler FPS-ów
clock = pygame.time.Clock()

# GRACZ
x = 200
y = 200
# Zmienne fizyczne
xVel = 0
yVel = 0

width = 200
height = 200
acceleration = 2

# Pętla gry
run = True
while run:

    # =============== ZAMYKANIE GRY ================= #

    for event in pygame.event.get():
        if event.type == QUIT:
            run = False
            break
        

    # =============== STEROWANIE ================= #    

    keys = pygame.key.get_pressed()

    if keys[K_d]:
        xVel += acceleration
    if keys[K_a]:
        xVel -= acceleration
    if keys[K_w]:
        yVel -= acceleration
    if keys[K_s]:
        yVel += acceleration

    yVel += 0.1

    xVel *= 0.98
    yVel *= 0.98

    x += xVel
    y += yVel

    
    # ======= OGRANICZENIE RUCHU GRACZA ======= #

    if x < 0:
        x = 0
        xVel *= -1

    if y < 0:
        y = 0
        yVel = 0

    if x > SCREEN_WIDTH - width:
        x = SCREEN_WIDTH - width
        xVel *= -1

    if y > SCREEN_HEIGHT - height:
        y = SCREEN_HEIGHT - height
        yVel = 0

    # =============== RENDEROWANIE ================= #
    # Wypełnianie ekranu kolorem
    screen.fill(COLOR_BLACK)

    # Rysowanie obiektu
    pygame.draw.rect(screen, COLOR_RED, Rect(x, y, width, height))
    pygame.draw.circle(screen, COLOR_WHITE, (x, y), 3)
    # Aktualizacja zawartości ekranu
    pygame.display.flip()

    # Czekanie 1/60 sekundy od poprzedniego wywołania clock.tick()
    clock.tick(FPS)

# Wyłączenie PyGame
pygame.quit()
```
